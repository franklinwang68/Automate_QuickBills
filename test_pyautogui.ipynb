### Procedures

1. Initial set up - vendor, total invoice amount, invoice date, bill invoice number, yes: create payable.
2. Enter Payables
    1. Enter and select related object on last row
    2. service date - month, day, year
    3. payable type
    4. amount
    5. edit payble - shift tab shit tab
    6. invoice number
    7. transaction date
    8. yes, pay in full
    9. save


### Code:
initial data
1. read csv file and parse dates
2. set transaction date

Enter payables
1. mouse click related object on last row (Wait before and after click).
    1. manual click, auto write, wait, down, tab, 
2. servmonth, tab, servday, tab, servyear, tab, 
3. payble type: pp (unless agreement) tab, tab
4. write amount tab enter
5. wait
6. shift tab shift tab wait
7. tab x times, write invoice, tab x4
8. transmonth, tab, transday, tab, transyear, tab
9. tab x times, down, enter


```python
import pyautogui
import time
import csv
```

```python
#Transaction Date
transactDay = '20'
transactMonth = 'Feb' #3 letter month
transactYear = '2025'
payable_type = "pat"  # set to "agreement" manually if needed

manual_click_target_row = True
```

```python
#Main Code

## about 50 records in 30 minutes

# === STEP 0: Initial Prep ===
print("Starting automation in 5 seconds — switch to the target window.")
time.sleep(5)

# === Load CSV ===
with open('data.csv', newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    
    for row in reader:
        # Extract fields from current CSV row
        case_number = row['Case Number']
        serv_month = row['Service Month']
        serv_day = row['Service Day']
        serv_year = row['Service Year']
        amount = row['Invoice Total']
        invoice = row['Invoice Number']

        print(f"Entering record: {invoice} ",end="")

        # === STEP 1: Go to target row ===
        if(manual_click_target_row):
            wait = 3
            print(f"Click target row within {wait} seconds.")
            time.sleep(wait)
        else:
            pyautogui.press('end')
            ### place cursor at target, don't move it
            time.sleep(2)
            pyautogui.click()

        print("Creating Payable... ", end="")
        # === STEP 2: Write case number ===
        pyautogui.write(case_number)
        time.sleep(5)
        pyautogui.press('down')
        pyautogui.press('tab')

        # === STEP 3: Service Date ===
        pyautogui.write(serv_month)
        pyautogui.press('tab')
        pyautogui.write(serv_day, interval=.5) 
        # In Sofia, pressing "1" and "1", the second "1" key acts like a down arrow. 
        # It will also do it whent the prior value is a single digit or between 10-19. 
        # This can be tested by entering "11" into the day field. The same occurs with "22"
        if (serv_day == '11'):
            pyautogui.press('down')
        if (serv_day == '22'):
            pyautogui.press('down')
            pyautogui.press('down')
        pyautogui.press('tab')
        pyautogui.write(serv_year)
        pyautogui.press('tab')

        # === STEP 4: Payable Type ===
        pyautogui.write(payable_type)
        pyautogui.press('tab')
        pyautogui.press('tab')

        # === STEP 5: Amount + Submit ===
        pyautogui.write(amount)
        pyautogui.press('tab')
        pyautogui.press('enter')

        print("Editing payable... ", end="")

        # === STEP 6: Edit Payable ===
        pyautogui.hotkey('shift', 'tab')
        pyautogui.hotkey('shift', 'tab')
        pyautogui.press('enter')
        time.sleep(7)

        # === STEP 7: Write invoice ===
        for _ in range(24):
            pyautogui.press('tab')
        pyautogui.write(invoice)

        # === STEP 8: Transaction Date ===
        for _ in range(4):
            pyautogui.press('tab')
        pyautogui.write(transactMonth)
        pyautogui.press('tab')

        ### Writing transact day
        # If day is 11 or 22 or single digit, need to reset it to default "- DAY -"
        for _ in range(31): 
           pyautogui.press('up')
        pyautogui.write(transactDay)

        if(transactDay == '11'): # special case 11
            pyautogui.press('down') 
        elif(transactDay == '22'): # special case 22
            pyautogui.press('down') 
            pyautogui.press('down') 
        
        pyautogui.press('tab')
        pyautogui.write(transactYear)
        pyautogui.press('tab')

        # === STEP 9: Yes, pay in full ===
        for _ in range(15):
            pyautogui.press('tab')
        pyautogui.press('down')
        pyautogui.press('enter')

        time.sleep(15)  # delay before next record
        print(f"✅ Finished: {invoice}")

print("✅ All records processed. Please Ensure all dates with 11 and 22 are entered correctly.")
```

```python
import requests

def check_page_status(url):
    try:
        response = requests.head(url, timeout=5)  # Use HEAD request for efficiency
        if response.status_code == 200:
            print(f"Page at {url} loaded successfully (status 200).")
            return True
        else:
            print(f"Page at {url} returned status code: {response.status_code}")
            return False
    except requests.exceptions.RequestException as e:
        print(f"Error checking page at {url}: {e}")
        return False

# Example usage:
check_page_status("https://case.wellspringsoftware.net/kms/bill/quick/10015/")
check_page_status("https://www.example.com/nonexistent-page")
```
